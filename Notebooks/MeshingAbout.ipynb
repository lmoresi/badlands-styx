# Meshing complex shapes with triangles

Here we use the `shapely` and `triangle` python packages to generate meshes from complex geometries.
These packages can be installed easily using `pip install ___`.

```python
%matplotlib inline

from badlands import meshtools, TreMesh, ThermMesh
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pylab import cm

minX, maxX, resX = 0.0, 50000.0, 1000
minY, maxY, resY = -35000.0, 0.0, 1000
xcoords = np.arange(minX, maxX+resX, resX)
ycoords = np.arange(minY, maxY+resY, resY)
```

Plain triangular mesh, no structure

```python
x1, y1, bmask, inverse_bmask = meshtools.square_mesh(minX, maxX, minY, maxY, resX, resY, 100, 20)
x1, y1 = meshtools.lloyd_mesh_improvment(x1, y1, bmask, 5)
mesh1  = TreMesh(x1,y1,bmask)

phi = x1+y1

fig = plt.figure(1)
ax = fig.add_subplot(111, xlim=[minX, maxX], ylim=[minY, maxY])

mesh1.add_node_data_to_plot(ax,phi, cmap=cm.terrain, shading="flat", vmin=-0.5, alpha=0.5)
ax.scatter(x1, y1, marker='o')

plt.show()
```

Lets create a geological boundary that we want to include. A `sin` curve will do.
We could add these points to the X, Y lists and remesh...

```python
Y = 1e4*np.sin(1e-4*xcoords) + minY/2

x2, y2 = np.concatenate([x1, xcoords]), np.concatenate([y1, Y])
bmask2 = np.concatenate([bmask, np.ones(len(xcoords), dtype=bool)])
x2, y2 = meshtools.lloyd_mesh_improvment(x2, y2, bmask2, 5)

mesh2  = TreMesh(x2,y2,bmask2)
phi2 = x2+y2

fig = plt.figure(1)
ax = fig.add_subplot(111, xlim=[minX, maxX], ylim=[minY, maxY])

mesh2.add_node_data_to_plot(ax,phi2, cmap=cm.terrain, shading="flat", vmin=-0.5, alpha=0.5)
ax.scatter(x2, y2, marker='o')
ax.plot(xcoords, Y, linewidth=2)

plt.show()
```

Not too bad with Lloyd's mesh improvement (hideous without!)
Ideally, it would be best to import structure and mesh using a specified number of points.

## Shapely

The *Shapely* package enables manipulation and computation of [complex geometries](http://www.toptal.com/python/computational-geometry-in-python-from-theory-to-implementation).
Below, we split the domain into 2 polygons and loop through each point to see whether it intersects one of them.

```python
import shapely.geometry

sineline = shapely.geometry.LineString(zip(xcoords, Y))
xc = np.concatenate(([minX], xcoords, [maxX]))
yc = np.concatenate(([minY], Y, [minY]))
bottomPolygon = shapely.geometry.Polygon(zip(xc, yc))
xc = np.concatenate(([minX], xcoords, [maxX]))
yc = np.concatenate(([maxY], Y, [maxY]))
topPolygon = shapely.geometry.Polygon(zip(xc, yc))
```

```python
# We can preview what it looks like!
topPolygon
```

```python
phi2 = np.zeros(len(x2), dtype=bool)
for i in range(len(x2)):
    pt = shapely.geometry.Point((x2[i], y2[i]))
    phi2[i] = pt.within(topPolygon) # Returns True/False

fig = plt.figure(1)
ax = fig.add_subplot(111, xlim=[minX, maxX], ylim=[minY, maxY])

mesh2.add_node_data_to_plot(ax,phi2, cmap=cm.terrain, shading="flat", vmin=-0.5, alpha=0.5)
ax.scatter(x2, y2, marker='o')
ax.plot(xcoords, Y, linewidth=2)

plt.show()
```

## Triangle

The `triangle` package is a 2D mesh generator using Delaunay triangulation.
It is a python binding based on Jonathan Richard Shewchuk’s [c-implementation](http://www.cs.cmu.edu/~quake/triangle.html).

```python
import triangle
import triangle.plot

# tri = triangle.delaunay(zip(x2, y2))
A = dict(vertices=np.array(zip(x2,y2)))
t = triangle.triangulate(A)

fig = plt.figure(1)
ax = fig.add_subplot(111, xlim=[minX, maxX], ylim=[minY, maxY])
triangle.plot.plot(ax, **t)
plt.show()
```

Now lets remesh the domain using just our top shape to create a constrained Delaunay triangulation

```python
def f7(seq):
    """ Removes duplicate entries while maintaining the same order """
    seen = set()
    seen_add = seen.add
    return [ x for x in seq if not (x in seen or seen_add(x))]

# We can simplify the number of vertices that define the shape
s1 = topPolygon.simplify(100, preserve_topology=False)
s2 = bottomPolygon.simplify(100, preserve_topology=False)

s1x, s1y = s1.exterior.coords.xy
vertices = f7(zip(s1x, s1y))
segments = zip(np.arange(0,len(vertices), dtype=int), np.append(np.arange(1, len(vertices), dtype=int), 0))


# Need to add segments to constrain the mesh boundary
B = dict(vertices = np.array(vertices), 
         segments = np.array(segments)
         )
```

The dictionary, `B`, is passed to the triangulation with a combination of options:

- `p` triangulates a Planar Straight Line Graph.
- `a` imposes a maximum area for each triangle.
- `q` quality mesh generation no angles smaller than specified degrees.
- `c` encloses convex hull with line segments

These are the main options, see the [API](http://dzhelil.info/triangle/) for a complete list. We create multiple subplots with different options below...

```python
t1 = triangle.triangulate(B, 'pq20') #a10000000
t2 = triangle.triangulate(B, 'pa'+str(resX*resY*5))
t3 = triangle.triangulate(B, 'p')

fig = plt.figure(1, figsize=(16,4))

ax1 = fig.add_subplot(131, xlim=[minX, maxX], ylim=[minY, maxY])
triangle.plot.plot(ax1, **t1)
ax2 = fig.add_subplot(132, xlim=[minX, maxX], ylim=[minY, maxY])
triangle.plot.plot(ax2, **t2)
ax3 = fig.add_subplot(133, xlim=[minX, maxX], ylim=[minY, maxY])
triangle.plot.plot(ax3, **t3)

plt.show()
```

The benefits of having a constrained mesh such as this are many. Here are a few obvious cases:

- A topographic surface or Moho surface can be defined.
- Less nodes are needed to resolve a geological contact = less computation time.
- Dynamic remeshing are useful for deformation mechanics like erosion or sediment transport.
